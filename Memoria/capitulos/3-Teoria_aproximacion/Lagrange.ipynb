# Ejemplo para ver cómo el error tiende a infinito al aproximar con polinomios
## Utiliza método de los polinomios de Lagrange 

```python
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt
from numpy.polynomial.polynomial import Polynomial
import numpy as np
```

```python
directorio_guardar_imagenes = "../../img/metodo-lagrange/"
for i in range(2,20,5):
    x = np.arange(-3, 3.1, 6/i)
    f = lambda z: np.log(z) if(z>1)  else np.exp(-z)-4
    y = list(map(f,x))
    poly = lagrange(x, y)
    x_new = np.arange(-3, 3.1, 0.1)
    plt.scatter(x, y, label='datos')
    plt.plot(x_new, Polynomial(poly.coef[::-1])(x_new), label='Polinomio Lagrange')
    plt.plot(x_new, list(map(f,x_new)) , label=r"$f(x)$", linestyle='-.')
    plt.legend()
    #plt.title(f"Polinomio lagrange para {i+1} datos")
    plt.savefig(directorio_guardar_imagenes+f"lagrange-{i+1}-datos.png")
    plt.show()
    
```
