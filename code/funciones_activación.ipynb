## Funci贸n umbral 

```julia
# Funci贸n umbral 
function CreateThreshold(polinomio, t)
    function ThresholdFunction(x)
        return  (polinomio(x) < t) ? 0 : 1
    end
    return ThresholdFunction
end

# Funci贸n lineal
p(x) = 2x 
CuadraticThreshold = CreateThreshold(p,1 )
display(plot( CuadraticThreshold, -1, 2, title="Funci贸n umbral p(x)= 2x, t=1", label=""))
display(plot( CuadraticThreshold, -1, 2, title="", label=""))
savefig(img_path*"funcion_umbral_lineal.png") 
```

```julia
function CosineSquasher(x::Real) :: Real
    salida = 1
    if x < - pi/2 
        salida =  0
    elseif x <=  pi/2 
        salida =  1/2 * (1 + cos( x - pi/2))
    end
    return salida
end

display(plot( CosineSquasher, -pi, pi, title="", label=""))
savefig(img_path*"cosineSquasherSinTitulo.png") 
```

## Funci贸n indicadora 

Es un caso particular de la funci贸n umbral

```julia
# Funci贸n indicadora 
function CreateIndicadora(t)
    function Funci贸nIndicadora(x)
        return  (x > t) ? 1 : 0
    end
    return Funci贸nIndicadora
end

indicadora = CreateIndicadora(0)
display(plot( indicadora, -1, 1, title="Funci贸n indicadora l = 0", label=""))
display(plot( indicadora, -1, 1, title="", label=""))

savefig(img_path*"funcion_indicadora_l_0.png") 
```

## Funci贸n rampa

```julia
function Rampa(x)
    salida = 0
    if 0<= x <=1
        salida = x
    elseif x>1
        salida = 1
    end
    return salida
end

display(plot( Rampa, -1, 2, title="Funci贸n rampa", label=""))
display(plot( Rampa, -1, 2, title="", label=""))

savefig(img_path*"funcion_rampa.png") 
```
